## Multivariate Plots

- We learned in previous lessons how to univariate and bivariate plots to plot data one or two variables at a time

- We used either position like is scatter plots or length like in bar charts

    - This is because our perception of values is most precise for these encodings
    
- This becomes an issue when we want to move on to multivariate visualizations (3 or more variables)

- We will learn about how to use encodings to add 3 or more variables to plots

## Non Positional Encodings for Third Variables

- We can use an encoding on a point on a scatter plot to indicate a third variable (say x is loan amount, y is interest rate, and the third value is average montly payment)

    - shape, size, or color
    
        - shape is best for categorical variables
        
            - e.g. Class of borrower
            
        - size is best for numeric variables
        
            - like monthly payment be of size 25, 50, or 100 value porportional to area
            
        - color: most common
        
            - it can be used for both quantitative or qualitative|
    

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv("./fuel_economy.csv")
# Ordinal by size of car
sedan_classes = [
 'Minicompact Cars','Subcompact Cars','Compact Cars',
 'Midsize Cars',
 'Large Cars']
vclasses =  pd.CategoricalDtype(ordered=True,
                                categories=sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)
fuel_econ['trans_type'] = \
    fuel_econ['trans'].apply(lambda x: x.split()[0])
```

```python
np.random.seed(2018)
sample = np.random.choice(fuel_econ.shape[0], 200, replace=False)
fuel_econ_subset = fuel_econ.loc[sample]
```

```python
ttype_markers = [["Automatic", 'o'],
                ["Manual", '^']]
for ttype, marker in ttype_markers:
    plot_data = \
        fuel_econ_subset.loc[fuel_econ_subset['trans_type'] == ttype]
    sb.regplot(data=plot_data, x='displ', y='comb',
          x_jitter=0.04, fit_reg=False,
          marker=marker);
plt.xlabel("Displacement (l)")
plt.ylabel("Combined Fuel Eff. (mpg)");
plt.legend(['Automatic', 'Manual']);
```

```python
sb.regplot(data=plot_data, x='displ', y='comb',
      x_jitter=0.04, fit_reg=False,
      scatter_kws ={'s': fuel_econ_subset['co2']/2});
plt.xlabel("Displacement (l)")
plt.ylabel("Combined Fuel Eff. (mpg)");

sizes = [200, 350, 500]
base_color = sb.color_palette()[0]
legend_obj = []
for s in sizes:
    legend_obj.append(plt.scatter([],[], s=s/2,color=base_color))
plt.legend(legend_obj, sizes, title="CO2 (g/mi)")
```

## Color Palettes

- After position, this is most important

- can be used for categorical and quantitative values

- we saw this in heat maps and clustered bar charts

- Note: saturation of a color is intensity of color

- qualitatitve palette

    - best for nominal data, distinct hues, similar brightness and saturation
    
     - no ordering is implied and no variable seems more important then others
 
 
- Sequential Pallets

    - used for ordinal and numeric data
    
        - light non saturated colors are lower values, higher more saturated colors on top (but we can flip this)
        
    - best to use a small range of hues or one hue
    
- divergent pallet

    - for ordinal or numeric data with meaningful center
    
    - Two sequential pallets with divergent hues put back to back with neutral center
    
    - one hue is greater then center, other is less then center
    

        
        

## After selecitng palette

- Think of color blindness

- particularlly important in qualitative and divergent scales

- Be mindful of using transparancy with color

    - if two points overlap, it gives us undefined new colors

```python
g = sb.FacetGrid(data=fuel_econ_subset, hue='trans_type',
                hue_order=["Automatic", "Manual"],
                size=4, aspect=1.5)
g.map(sb.regplot, 'displ', 'comb',  x_jitter=0.04, fit_reg=False)
g.add_legend()
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
g = sb.FacetGrid(data=fuel_econ_subset, hue='VClass',
                size=4, aspect=1.5,
                palette='viridis_r')
g.map(sb.regplot, 'displ', 'comb',  x_jitter=0.04, fit_reg=False)
g.add_legend()
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
plt.scatter(data=fuel_econ_subset, x="displ",  y="comb", c = "co2")
plt.colorbar(label="CO2 (g/mi)")
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
sample = np.random.choice(pokemon.shape[0], 250, replace=False)
pokemon_subset = pokemon.loc[sample]
```

```python
plt.scatter(data=pokemon_subset, x="defense", 
            y="special-defense", c = "speed")
plt.colorbar(label="Speed")
plt.xlabel('Defense')
plt.ylabel('Sp. Defense');
# m, b = np.polyfit(pokemon_subset["defense"],
#                   pokemon_subset["special-defense"],
#                  1)
# plt.plot(pokemon_subset["defense"], m * pokemon_subset["defense"] + b)
```

```python
type_cols = ['type_1','type_2']
non_type_cols = pokemon.columns.difference(type_cols)
pkmn_types = pokemon.melt(id_vars = non_type_cols, value_vars = type_cols, 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

```python
dragon_types = pkmn_types.query("type == 'dragon'")
fairy_types = pkmn_types.query("type == 'fairy'")
types = pd.concat([dragon_types, fairy_types]).reset_index()
types.sample(10)
```

```python
g = sb.FacetGrid(data=types, hue='type',
                size=4, aspect=1.5,
                )
g.map(sb.regplot, 'weight', 'height',  x_jitter=0.04, fit_reg=False)
g.add_legend()
plt.xlabel('Weight kg')
plt.ylabel('Height m');
```

## Faceting in two directions

- We don't only have to use additional encodings

- we can use facetting of bivariate plots

- We can choose to either make rows and columns and facet in two directions (more univariate plots)

```python
q = sb.FacetGrid(data=fuel_econ, col="VClass", row="trans_type",
                margin_titles=True)
q.map(plt.scatter, 'displ', 'comb');
```

## Other adaptions of bivariate plots

- We can change count access to mean of second variable.

    - clustered bar charts and heat maps, we can do the same approach to make them multivariate
    
- In general, we can take an aggregated third variable and plot it in a bivariate plot. This is the same tequnique we used with univariate plots

```python
sb.barplot(data=fuel_econ, x='VClass', y='comb',
            hue='trans_type', ci='sd');
plt.xticks(rotation=15);
plt.ylabel("Avg combined efficiency (mpg)");
```

```python
sb.boxplot(data=fuel_econ, x='VClass', y='comb',
            hue='trans_type');
plt.xticks(rotation=15);
plt.ylabel("Avg combined efficiency (mpg)");
```

## Plot matrices

- We can explore relationships to lots of variables in bivariate exploration

- In a plot matrix, we generate a grid of. We have feature by feature and generate scatter plots, except in diagonal where we have univariate plots

- It is best for an initial exploratory tool, because the plots are typically very small

- We should then move to other viz after using plot matrix

- We can also make a heat map with corrlations

    - good summary option to start with



```python
pkmn_stats = ['hp', 'attack', 'defense', 'speed',
       'special-attack', 'special-defense']
g = sb.PairGrid(data=pokemon, vars=pkmn_stats)
g= g.map_offdiag(plt.scatter)
g.map_diag(plt.hist)
```

```python
sb.heatmap(pokemon[pkmn_stats].corr(), cmap='rocket_r', annot=True,
          vmin=0);
```

Note that seeing 0.51 for defense and sp defense and seeing outlier can give us clue that outlier is present
